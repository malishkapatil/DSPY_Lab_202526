####1. Need for Hypothesis Testing

Hypothesis testing is a statistical technique used to make decisions about a population based on sample data.

####Need and Importance:

1. **Decision Making**
   - Helps in accepting or rejecting claims about population parameters.

2. **Scientific Research**
   - Used to validate theories, assumptions, and models.

3. **Handling Uncertainty**
   - Since complete population data is not available, sampling is used.

4. **Distinguishing Real Effects**
   - Determines whether results are due to actual effects or random chance.

5. **Comparison of Groups**
   - Used to compare means (e.g., two classes, treatments).

6. **Quality Control**
   - Used in industries to maintain standards.

Testing whether the average age of students in TE IT class is 19 years.

##**2.With a neat diagram show important steps in hypothesis Testing.**

####**Ans:**

**Steps in Hypothesis Testing**

**State the Hypotheses**

* Null Hypothesis (H₀): Assumes no effect or no difference.

* Alternative Hypothesis (H₁): Represents the effect or difference you want to test.

**Choose the Significance Level (α)**

* Commonly set at 0.05 (5%).

* Defines the threshold for rejecting H₀.

**Select the Appropriate Test**

* Based on data type, sample size, and distribution (e.g., t-test, chi-square test, ANOVA).

**Collect and Summarize Data**

* Gather sample data.

* Compute test statistics (e.g., t-value, z-value).

**Calculate the p-value or Critical Value**

* p-value: Probability of observing the data if H₀ is true.

* Compare test statistic with critical value from distribution tables.

**Make a Decision**

* If p-value ≤ α → Reject H₀ (evidence supports H₁).

* If p-value > α → Fail to reject H₀ (insufficient evidence against H₀).

**Draw a Conclusion**

* Interpret results in the context of the research question.

* Report findings clearly

##**3.What is a student t distribution?**

####**Ans:**

**Student t-Distribution**

The Student's t-distribution is used when:
- Sample size is small (n < 30)
- Population standard deviation is unknown

#### Characteristics:
- Symmetrical and bell-shaped
- Has heavier tails than normal distribution
- Depends on degrees of freedom (df = n − 1)
- Approaches normal distribution as sample size increases

#### Importance:
- Used in t-tests
- Provides more accurate results for small samples


##**4.What is level of significance ? What role t plays in hypothesis Testing.**

####**Ans:**

**Level of Significance and Role of t in Hypothesis Testing**

The **level of significance**, denoted by α (alpha), is the probability threshold used in hypothesis testing to decide whether to reject the null hypothesis. It represents the **maximum risk of making a Type I error**, which occurs when the null hypothesis is wrongly rejected even though it is true. Commonly used significance levels are 0.05, 0.01, and 0.10.  

The **t-statistic** plays a central role in hypothesis testing when the population standard deviation is unknown and the sample size is small. It measures how far the sample mean deviates from the hypothesized population mean in terms of standard errors. This t-value is then compared against the **critical value from the Student’s t-distribution** corresponding to the chosen level of significance and degrees of freedom. If the t-statistic falls in the critical region, the null hypothesis is rejected; otherwise, it is not rejected. Thus, the t-statistic provides a quantitative basis for making decisions in hypothesis testing.


##**5.Design a hypothesis testing for identifying average age of students in TE IT class is 19 or not.**

####**Ans:**

**Hypothesis Testing: Average Age of Students in TE IT Class**

**Objective:** To determine whether the average age of students in the TE IT class is 19 years.

**Step 1: Define Hypotheses**  
- **Null Hypothesis (H₀):** The average age of students in the TE IT class is 19 years. (μ = 19)  
- **Alternative Hypothesis (H₁):** The average age of students in the TE IT class is not 19 years. (μ ≠ 19)  

**Step 2: Select Significance Level**  
- Choose a **level of significance (α)**, typically 0.05.

**Step 3: Collect Sample Data**  
- Collect a random sample of students’ ages from the TE IT class.

**Step 4: Calculate Test Statistic**  
- Use the **t-test** since the population standard deviation is unknown and the sample size is likely small.  
- Formula for t-statistic:  
\[
t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}}
\]  
Where:  
  - \( \bar{x} \) = sample mean  
  - \( \mu_0 \) = hypothesized mean (19)  
  - \( s \) = sample standard deviation  
  - \( n \) = sample size  

**Step 5: Determine Critical Value / p-value**  
- Compare the calculated t-statistic with the **critical t-value** from the Student’s t-distribution table at α level and df = n-1.  
- Alternatively, compute the **p-value**.

**Step 6: Make Decision**  
- If |t-statistic| > t-critical or p-value < α → Reject H₀ (average age is not 19).  
- Otherwise → Fail to reject H₀ (insufficient evidence to conclude the average age differs from 19).  

**Step 7: Conclusion**  
- State the result in context: whether the data provides enough evidence to support that the average age of students is 19 or not.

**3. Python Implementation**

Here’s a ready-to-run Python snippet for your notebook:

```python
import numpy as np
from scipy import stats

# Sample data: ages of TE IT students
ages = np.array([19, 20, 18, 19, 20, 19, 18, 19, 20, 19])  # replace with your data

# Hypothesized mean
mu_0 = 19

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(ages, mu_0)

# Significance level
alpha = 0.05

print("Sample mean:", np.mean(ages))
print("t-statistic:", t_stat)
print("p-value:", p_value)

# Decision based on p-value
if p_value > alpha:
    print("Fail to reject H0: There is not enough evidence to say the average age is different from 19.")
else:
    print("Reject H0: The average age is significantly different from 19.")
```
